```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from feature_creation import df_features as df
```

```python
num_solutions = df["NUM_SOLUTIONS"]
print(num_solutions.describe())

sns.histplot(num_solutions)
```

```python
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmin = -1, vmax=1, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
corr_solutions = corr.NUM_SOLUTIONS.reset_index()
corr_solutions = corr_solutions[1:]
corr_solutions.sort_values(by = "NUM_SOLUTIONS", inplace = True)

fig_dims = (4, 18)
fig, ax = plt.subplots(figsize=fig_dims)

corr_chart = sns.barplot(y="index", x="NUM_SOLUTIONS", data=corr_solutions, ax = ax)
```
